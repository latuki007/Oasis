```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
car_data = pd.read_csv('car data.csv')
```

```python
print(car_data.Transmission.value_counts())
```

```python
car_data.replace({'Fuel_Type':{'Petrol':0, 'Diesel':1, 'CNG':2}}, inplace = True)
```

```python
car_data.replace({'Selling_type':{'Dealer':0, 'Individual':1}}, inplace = True)
```

```python
car_data.replace({'Transmission':{'Automatic':0, 'Manual':1, }}, inplace = True)
```

```python
car_data.head()
```

```python
x = car_data.drop(columns = ['Car_Name', 'Selling_Price'])
```

```python
print(x)
```

```python
y = car_data['Selling_Price']
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.1, random_state = 2)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
training_prediction = model.predict(x_train)
```

```python
from sklearn import metrics
```

```python
error_score = metrics.r2_score( y_train, training_prediction)
```

```python
print("R^2 Error:", error_score)
```

```python
plt.scatter( y_train, training_prediction)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual vs Predicted Prices")
```

```python
test_prediction = model.predict(x_test)
```

```python
error_score = metrics.r2_score(y_test,test_prediction)
```

```python
print("R^2 Error:", error_score)
```

```python
plt.scatter( y_test, test_prediction)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual vs Predicted Prices")
```
